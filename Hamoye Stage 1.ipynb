```python
#Num & Percentages of missing values
missing_nums = food_balance_sheet["Y2014"].isna().sum()
missing_nums
```

```python
#percentage of missing nums
total_nums = food_balance_sheet["Y2014"].shape[0]
percentage = missing_nums / total_nums * 100
percentage
```

```python
#groupby operation on 'Element'
element_sum = food_balance_sheet.groupby("Element")[["Y2014", "Y2015", "Y2016", "Y2017", "Y2018"]].sum()
element_sum
```

```python
#Highest Sum of Stock Variation
highest_stock_variation = element_sum[element_sum.index == "Stock Variation"].max()
highest_stock_variation
```

```python
#Total num of the sum of Processing in 2017
processing_sum = element_sum[element_sum.index == "Processing"]
processing_sum
```

```python
processing_sum[["Y2014", "Y2015", "Y2016", "Y2017", "Y2018"]].sum(axis=1)
```

```python
#Area with the highest sum in 2017
area_2017 = food_balance_sheet[["Y2017", "Area"]]
area_2017.head()
```

```python
area_2017.groupby("Area")[["Y2017"]].sum().sort_values(by="Y2017", ascending=False).head(1)
```

```python
#Area with 7th lowest sum in 2017
area_2017.groupby("Area")[["Y2017"]].sum().sort_values(by="Y2017", ascending=True).head(7)
```
